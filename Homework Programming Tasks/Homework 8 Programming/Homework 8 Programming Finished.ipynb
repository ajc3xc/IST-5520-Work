```python
import pandas as pd
import statsmodels.api as sm
from patsy import dmatrices

df = pd.read_csv('UniversalBank.csv')
df.head()
```

```python
#Get dummies
df = pd.get_dummies(df, columns=['Education'])

df.head(3)
```

```python
formula = 'Personal_Loan ~ Age + Experience + Income + Family + CCAvg + Education_1 + Education_2 + Education_3 + Mortgage + Securities_Account + CD_Account + Online + CreditCard'

y, X = dmatrices(formula, data = df, return_type='dataframe')

y.head()
```

```python
mod = sm.OLS(y, X)

res = mod.fit()

print(res.summary())
```
